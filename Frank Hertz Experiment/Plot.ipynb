```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Table 1: U_F = 8V, U_KG = 6V, U_AG = 5V


```python
U = pd.read_csv('TABLE 1.csv')
U1 = U['Ua (V)']
I1 = U['I (nA)']
plt.plot(U1, I1, 'x-')
for x in [18, 36]:  
    plt.axvline(x=x, color='r', linestyle='--')
plt.xlabel('Accelarating voltage (U) (V)')
plt.ylabel('Collector Current (I) (nA)')
plt.title('Frank Hertz Experiment Table 1')
plt.grid()
plt.show()
```
