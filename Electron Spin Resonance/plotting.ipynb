Magnetic Resonance using Compass

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Compass Experiment Magnetic field vs Resonant frequency 

```python
from scipy.optimize import curve_fit

D = pd.read_csv('magnetic_resonance_compass.csv')
f = D['Frequency (Hz)'].values 
B = D['Magnetic field (mT)'].values
# Fit f as a function of sqrt(B)

def sqrt_fit(x, a, b):
    return a * np.sqrt(x) + b

params, _ = curve_fit(sqrt_fit, B, f)
f_sqrt_fit = sqrt_fit(B, *params)

plt.plot(B, f_sqrt_fit, color='purple', label='f = a*sqrt(B) + b fit')
plt.scatter(B, f, color='orange', label='Data (f vs B)')
plt.xlabel('Magnetic field (mT)')
plt.ylabel('Frequency (Hz)')
plt.title('Frequency vs Magnetic field')
plt.legend()
plt.grid()
plt.show()
print('Fit parameters (a, b):', params)
```
